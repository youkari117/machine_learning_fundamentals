```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
print(cancer["DESCR"])
```

```python
cancer_data = pd.DataFrame(cancer["data"], columns = cancer["feature_names"])
```

```python
cancer_data
```

```python
cancer_target = pd.DataFrame(cancer["target"], columns = ["diagnostic"])
```

```python
cancer_target
```

```python
cancer["target_names"]
```

```python
cancer_df = pd.concat([cancer_data, cancer_target], axis = 1)
```
