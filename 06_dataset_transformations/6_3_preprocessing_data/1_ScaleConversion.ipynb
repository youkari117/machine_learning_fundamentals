<h1>Summary</h1>

<h4>StandardScaler</h4>
個々の特徴量の平均が0で分散が1になるように変換し、すべての特徴量の大きさを揃える<br>
しかし、この方法は特徴量の最大値や最小値がある範囲に入ることを保証するものではない<br>
<h4>RobustScaler</h4>
RobustScalerは、StandardScalerと異なり、平均値と標準偏差の代わりに中央値と四分位範囲を用いるため、外れ値を無視する<br>
外れ値は他のスケール変換方法では問題になる<br>
<h4>MinMaxScaler</h4>
データがちょうど0から1の間に入るように変換する<br>
訓練データの最小値を引き、訓練データのレンジ（最大 - 最小）で割る<br>
<h4>Normalizer</h4>
個々のデータポイントを、特徴量ベクトルがユークリッド長1になるように変換する。言い換えると、データポイントを半径1の円（より高次元なら超球面）に投射する。したがって、全てのデータポイントに対してそれぞれ異なるスケール変換が行われる（もとのユークリッド長の逆数をかける）<br>
この変換は、特徴ベクトルの長さではなく、方向（もしくは角度）だけが問題になる場合に用いられる

<h1>MinMaxScaler</h1>

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state = 1)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(C = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler().fit(X_train)
```

```python
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svm.fit(X_train_scaled, y_train)
```

```python
print(f"SVM test accuracy: {svm.score(X_test_scaled, y_test):.2f}")
```
